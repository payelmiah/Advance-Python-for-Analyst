Join Sets
There are several ways to join two or more sets in Python.

The union() and update() methods joins all items from both sets.

The intersection() method keeps ONLY the duplicates.

The difference() method keeps the items from the first set that are not in the other set(s).

The symmetric_difference() method keeps all items EXCEPT the duplicates.

**Union**

```python
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}

set3 = set1.union(set2)
print(set3)
```

Symmetric Differences
The symmetric_difference() method will keep only the elements that are NOT present in both sets.

```python
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.symmetric_difference(set2)

print(set3)
```

```python
#You can use the ^ operator instead of the symmetric_difference() method, and you will get the same result.
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1 ^ set2
print(set3)
```

```python
#Use the symmetric_difference_update() method to keep the items that are not present in both sets:

set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set1.symmetric_difference_update(set2)

print(set1)
```

# **Python Decorators**
# A decorator is a function that takes another function and returns a new function.


# @changecase means exactly the same as: myfunction = changecase(myfunction)

```python
# This is a decorator function
# It takes another function as an argument
def changecase(func):

    # This is the inner (wrapper) function
    # It adds extra behavior to the original function
    def myinner():
        # Call the original function
        result = func()

        # Convert the returned text to uppercase
        return result.upper()

    # Return the inner function (NOT calling it)
    return myinner


# Applying the decorator to a function
# This means: myfunction = changecase(myfunction)
@changecase
def myfunction():
    # Original function returns a simple string
    return "Hello Sally"


# Calling the decorated function
# Actually, this calls myinner(), not the original myfunction()
print(myfunction())
```

```python
def decoratorfuncion(func):
  def innerfunction():
    print("This is the decorator")
    result=func()
    return result.upper()
  return innerfunction


@decoratorfuncion
def mainfunction():
  return "Hello world"

print(mainfunction())
```

```python
def changecase(func):
  def myinner():
    return func().upper()
  return myinner

@changecase
def myfunction():
  return "Hello Sally"

@changecase
def otherfunction():
  return "I am speed!"

print(myfunction())
print(otherfunction())
```

Arguments in the Decorated Function

```python

def changecase(func):
  def myinner(name):
    return func(name).upper()
  return myinner

@changecase
def myfunction(nam):
  return "hello " + nam
print(myfunction("Emil"))
```

```python
def changecase(n):
  def changecase(func):
    def myinner():
      if n == 1:
        a = func().lower()
      else:
        a = func().upper()
      return a
    return myinner
  return changecase

@changecase(1)
def myfunction():
  return "Hello Linus"

print(myfunction())
```

#*args and **kwargs

```python
def changecase(func):
  def myinner(*args, **kwargs):
    return func(*args, **kwargs).upper()
  return myinner

@changecase
def myfunction(nam):
  return "Hello " + nam

print(myfunction("John"))
```

```python
def changecase(func):
  def myinner():
    return func().upper()
  return myinner

def addgreeting(func):
  def myinner():
    return "Hello " + func() + " Have a good day!"
  return myinner

@changecase
@addgreeting
def myfunction():
  return "Tobias"

print(myfunction())
```

```python
import functools

def changecase(func):
  @functools.wraps(func)
  def myinner():
    return func().upper()
  return myinner

@changecase
def myfunction():
  return "Have a great day!"

print(myfunction.__name__)
```

```python
# using decorator you can use @property and @age.setter instead of writing get_age and set_age

class Person:
  def __init__(self, name, age):
    self.name = name
    self.__age = age # private property

  @property
  def age(self): # use a getter method to access a private property
    return self.__age

  @age.setter
  def age(self, age): # use a setter method to modify a private property
    self.__age = age if 0 <= age <= 120 else print("Invalid age")

obj = Person(input("Enter name: "), int(input("Enter age: ")))
print(obj.name)
print(obj.age)
```

```python
# =============================
# Base Class
# =============================
class LibraryItem:
    def __init__(self, title):
        # Private attribute (encapsulation)
        self.__title = title

        # Creating an inner class object
        self.item_id = self.ItemID(1001)

    # -------------------------
    # Getter (read access)
    # -------------------------
    @property
    def title(self):
        return self.__title

    # -------------------------
    # Setter (write access)
    # -------------------------
    @title.setter
    def title(self, title):
        if title.strip() == "":
            print("Invalid title")
        else:
            self.__title = title

    # -------------------------
    # Private method
    # -------------------------
    def __log_access(self):
        print("Access logged for item")

    # -------------------------
    # Method to be overridden
    # -------------------------
    def display_info(self):
        print(f"Title: {self.title}")
        self.__log_access()
        self.item_id.show_id()

    # =============================
    # Inner Class
    # =============================
    class ItemID:
        def __init__(self, id_number):
            self.__id_number = id_number

        def show_id(self):
            print(f"Item ID: {self.__id_number}")


# =============================
# Child Class: Book
# =============================
class Book(LibraryItem):
    def __init__(self, title, author):
        # Call parent constructor
        super().__init__(title)
        self.__author = author

    @property
    def author(self):
        return self.__author

    # Method overriding (polymorphism)
    def display_info(self):
        super().display_info()
        print(f"Author: {self.__author}")


# =============================
# Child Class: Magazine
# =============================
class Magazine(LibraryItem):
    def __init__(self, title, issue_number):
        super().__init__(title)
        self.__issue_number = issue_number

    @property
    def issue_number(self):
        return self.__issue_number

    # Method overriding (polymorphism)
    def display_info(self):
        super().display_info()
        print(f"Issue Number: {self.__issue_number}")


# =============================
# Polymorphism in action
# =============================
items = [
    LibraryItem("Programming Language"),
    Book("Python", "John"),
    Magazine("BBC", 1)
]

for item in items:
    item.display_info()
    print()
```

```python
class Payment:
  def __init__(self, amount):
    self.__amount = amount
    self.transaction_id = self.Transaction(1001)

  @property
  def amount(self):
    return self.__amount

  @amount.setter
  def amount(self, amount):
    if amount <= 0:
      raise ValueError("Invalid amount")
    else:
      self.__amount = amount

  def process_payment(self):
    print("Processing payment...")
    self.__validate_payment()
    self.transaction_id.show_transaction()

  def __validate_payment(self):
    if self.__amount > 0:
      print("Payment validated")
    else:
      print("Invalid payment amount")


  class Transaction:
    def __init__(self, transaction_id):
      self.__transaction_id = transaction_id

    def show_transaction(self):
      print(f"Transaction ID: {self.__transaction_id}")

class CreditCardPayment(Payment):
  def __init__(self, amount, card_number):
    super().__init__(amount)
    self.__card_number = card_number


  @property
  def card_number(self):
    return self.__card_number

  def process_payment(self):
    super().process_payment()
    print("Credit card payment processed")

class PayPalPayment(Payment):
  def __init__(self, amount, email):
    super().__init__(amount)
    self.__email = email

  @property
  def email(self):
    return self.__email

  def process_payment(self):
    super().process_payment()
    print("PayPal payment processed")



polylist = [
    Payment(100),
  CreditCardPayment(200, "1234567890"),
    PayPalPayment(300, "william@gmail.com")
]

for payment in polylist:
  payment.process_payment()
  print()
```

**Python File Write**

To write to an existing file, you must add a parameter to the open() function:

"a" - Append - will append to the end of the file

"w" - Write - will overwrite any existing content

```python

with open("demofile.txt", "a") as f:
  f.write("Now the file has more content!")

#open and read the file after the appending:
with open("demofile.txt") as f:
  print(f.read())
```
